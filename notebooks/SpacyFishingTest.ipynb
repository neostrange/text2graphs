```python
import requests

def amuse_wsd_api_call(api_endpoint, sentences):
    headers = {
        "accept": "application/json",
        "Content-Type": "application/json"
    }
    data = [{"text": sent, "lang": "EN"} for sent in sentences]
    data_json = "[" + ",".join([f'{{"text": "{item["text"]}", "lang": "{item["lang"]}"}}' for item in data]) + "]"

    try:
        response = requests.post(api_endpoint, data=data_json, headers=headers)
        response.raise_for_status()
        return response.json()
    except requests.exceptions.RequestException as e:
        print(f"Error while calling AMuSE-WSD API: {e}")
        return None

if __name__ == "__main__":
    # Define the AMuSE-WSD API endpoint
    amuze_wsd_api_endpoint = "http://localhost:81/api/model"

    # Example list of sentences
    sentences = [
        "He went to Lahore last night",
        "She visited the museum in the morning"
    ]

    # Call the AMuSE-WSD API
    api_response = amuse_wsd_api_call(amuze_wsd_api_endpoint, sentences)

    if api_response:
        print(api_response)
    else:
        print("API call failed.")
```

```python
from nltk.corpus import wordnet31

# Assuming you have a token with wnSynsetOffset, e.g., '15167027n'
wn_synset_offset = '1568684a'

# Convert the wnSynsetOffset to an integer (if it's stored as a string)
offset = int(wn_synset_offset[:-1])  # Convert '15167027n' to integer 15167027
pos = wn_synset_offset[-1]  # Get the part-of-speech from the last character 'n'

# Retrieve the synset using synset_from_pos_and_offset
synset = wordnet31.synset_from_pos_and_offset(pos, offset)

# Function to get all hypernyms recursively
def get_all_hypernyms(synset):
    hypernyms = []
    hypernym_synsets = synset.hypernyms()
    for hypernym_synset in hypernym_synsets:
        hypernyms.append(hypernym_synset)
        hypernyms.extend(get_all_hypernyms(hypernym_synset))  # Recursive call to get hypernyms of hypernyms
    return hypernyms

# Get all hypernyms of the synset
all_hypernyms = get_all_hypernyms(synset)

# Print the synset properties and all hypernyms
print("Synset definition:", synset.definition())
print("Synset examples:", synset.examples())
print("Synset hypernyms:", all_hypernyms)
print("Synset hyponyms:", synset.hyponyms())
print("Synset lemmas:", synset.lemmas())
```

```python
from nltk.corpus import wordnet

# Assuming you have a token with wnSynsetOffset, e.g., '15167027n'
wn_synset_offset = '15167027n'

# Convert the wnSynsetOffset to an integer (if it's stored as a string)
offset = int(wn_synset_offset[:-1])  # Convert '15167027n' to integer 15167027
pos = wn_synset_offset[-1]  # Get the part-of-speech from the last character 'n'

# Retrieve the synset using synset_from_pos_and_offset
synset = wordnet.synset_from_pos_and_offset(pos, offset)

# Function to get all hypernyms recursively
def get_all_hypernyms(synset):
    hypernyms = []
    hypernym_synsets = synset.hypernyms()
    for hypernym_synset in hypernym_synsets:
        hypernyms.append(hypernym_synset)
        hypernyms.extend(get_all_hypernyms(hypernym_synset))  # Recursive call to get hypernyms of hypernyms
    return hypernyms

# Get all hypernyms of the synset
all_hypernyms = get_all_hypernyms(synset)

# Function to get the domain labels for a synset
def get_domain_labels(synset):
    domain_labels = []
    topic_domains = synset.topic_domains()
    for domain in topic_domains:
        domain_labels.extend(domain.split("."))
    return domain_labels

# Get domain labels for the synset
domain_labels = get_domain_labels(synset)

# Function to get synonyms of a synset
def get_synonyms(synset):
    synonyms = []
    for lemma in synset.lemmas():
        synonyms.append(lemma.name())
    return synonyms

# Get synonyms for the synset
synonyms = get_synonyms(synset)

# Print the synset properties, hypernyms, domain labels, and synonyms
print("Synset definition:", synset.definition())
print("Synset examples:", synset.examples())
print("Synset hypernyms:", all_hypernyms)
print("Synset hyponyms:", synset.hyponyms())
print("Synset lemmas:", synset.lemmas())
print("Synset domain labels:", domain_labels)
print("Synset synonyms:", synonyms)
```
